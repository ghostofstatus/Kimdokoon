```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./csv/data_0917_20220716.csv', encoding='CP949')
```

```python
df.head()
```

```python
df1 = df.copy()
df1['위험도'] = 0
for i in range(len(df1)) :
    if df1['공매도거래대금증가율'][i] >= 6.0 :
        df1['위험도'][i] += 3
    elif df1['공매도거래대금증가율'][i] >= 4.0 :
        df1['위험도'][i] += 2
    elif df1['공매도 거래대금 증가배율'][i] >= 2.0 : 
        df1['위험도'][i] += 1

```

```python
for i in range(len(df1)) :
    if df1['주가 수익률'][i] <= -10.0 :
        df1['위험도'][i] += 4
    elif df1['주가 수익률'][i] <= -5.0 :
        df1['위험도'][i] += 2
    elif df1['주가 수익률'][i] <= -3.0: 
        df1['위험도'][i] += 1
```

```python
df1['위험도'][[0,1,2,3,4]] += 3
df1['위험도'][[5,6,7,8,9]] += 2
df1['위험도'][[10,11,12,13,14]] += 1
```
