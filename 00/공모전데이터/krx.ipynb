## 공매도 krx 함수
### krx.pykrx
krx.pykrx(data, value, n, str, end)

    data : 기준(띄어쓰기없이 다 붙여써야 함수가 돌아감)
        1 : 종목별 공매도 현황
        2 : 종목별 공매도 거래 정보
        3 : 일자별 공매도 거래 현황
        4 : 투자자별 공매도 거래량 현황
        5 : 투자자별 공매도 거래대금 현황
        6 : 종목별공매도잔고현황
    * value : 기준안의 컬럼값(입력하면 그래프가 하나의 피쳐만 갖고 나옴)
    n : 종목코드
    str : 시작날짜, 확인할 날짜
    end : 끝날짜
    필요없는 부분은 '' 공백으로 주면 됨


    1. get_shorting_status_by_date('str', 'end', 'n')
        종목별 공매도 현황
        컬럼 : 거래량 , 잔고수량 , 거래대금 , 잔고금액

    2. get_shorting_volume_by_ticker('str')
        종목별 공매도 거래 정보
        컬럼 : 공매도 , 매수 , 비중

    3. get_shorting_volume_by_date('str', 'end', 'n')
        일자별공매도거래현황
        컬럼 : 공매도, 매수, 비중

    4. get_shorting_investor_volume_by_date('str', 'end', 'n')
        투자자별공매도거래량현황
        컬럼 : 기관, 개인, 외국인, 기타, 합계
    
    5. get_shorting_investor_value_by_date('str', 'end', 'n')
        투자자별공매도거래대금현황
        컬럼 : 기관, 개인, 외국인, 기타, 합계

    6. get_shorting_balance_by_date('str' , 'end', 'n')
        종목별공매도잔고현황
        컬럼 : 공매도잔고, 상장주식수, 공매도금액, 시가총액, 비중

### krx.rank
krx.rank(data, day)

    data : 기준(띄어쓰기없이 다 붙여써야 함수가 돌아감)
        1 : 공매도비중 상위 50
        2 : 공매도잔고 비중 50
    day : 확인할 날짜

    1. get_shorting_volume_top50(day)
        공매도 비중 상위 50
        컬럼 : 순위, 공매도거래대금, 총거래대금, 공매도비중, 직전40일거래대금평균
    
    2. get_shorting_balance_top50(day)
        공매도 잔고 상위 50
