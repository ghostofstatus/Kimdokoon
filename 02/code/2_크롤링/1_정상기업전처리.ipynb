```python
import pandas as pd

df = pd.read_excel('/Users/ijunhui/Downloads/ubion/프로젝트2/1_기업선정/정상기업/2017정상기업.xlsx')
df['시간'] = pd.to_datetime(df['시간'])
# df = df.drop('번호', axis=True)
df = df.sort_values('시간')
df = df.set_index(keys='시간')
df = df.reset_index()
df = df.drop_duplicates(['회사명'], keep = 'first') #관리종목 중복 지정 시 가장 먼저 지정된 시점만 남기고 중복 데이터 삭제
df['시간'] = df['시간'].dt.strftime("%Y.%m.%d")
df['시간'] = pd.to_datetime(df['시간'])
df = df.set_index(keys='시간')
df = df.reset_index()
df
```

```python
from dateutil.relativedelta import relativedelta

#6개월 전 데이터 추출 위해 날짜 계산

new = []
for i in range(0, len(df)):
    n = df.loc[i]['시간'] - relativedelta(months=6)
    new.append(n)
```

```python
df['기간'] = pd.DataFrame(new)
```

```python
df = df[['시간', '회사명','기간']]
```
