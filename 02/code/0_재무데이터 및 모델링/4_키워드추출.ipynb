## 뉴스데이터 키워드 추출

#### 관리기업 뉴스데이터 read

```python
import pandas as pd

df = pd.read_csv('관리기업크롤링전처리.csv')
df
```

#### 뉴스데이터 제목과 본문의 내용 결합 -> df['전체']

```python
df['전체'] = df['제목'] + " " + df['본문']

df
```

#### 불용어 처리


```python
import re

## () [] 안의 단어는 기자 이름 등 불필요한 단어이므로 제거하고
## 불필요한 단어 특수문자 제거
pattern1 = r'\([^)]*\)'
pattern2 = r"\[([^]]+)\]"

df['뉴스기사본문전처리'] = [re.sub(pattern1, '', s) for s in df['전체']]
df['뉴스기사본문전처리'] = [re.sub(pattern2, '', s) for s in df['뉴스기사본문전처리']]
df['뉴스기사본문전처리'] = [re.sub('[^/^$|\s+/가-힣\.]', '', s) for s in df['뉴스기사본문전처리']]
df['뉴스기사본문전처리'] = [re.sub('[[ \s]{2,}\.{2,}]', '', s) for s in df['뉴스기사본문전처리']]
df['뉴스기사본문전처리'] = [re.sub('\n', '', s) for s in df['뉴스기사본문전처리']]
df['뉴스기사본문전처리'] = [re.sub('[/+]', '', s) for s in df['뉴스기사본문전처리']]
df['뉴스기사본문전처리'] = [re.sub('\r', '', s) for s in df['뉴스기사본문전처리']]

df['뉴스기사본문전처리']
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import urllib.request
from gensim.models.word2vec import Word2Vec
from konlpy.tag import Okt
import konlpy
okt = Okt()
```

#### Null값 제거

```python
train_data = df['뉴스기사본문전처리'].dropna(how = 'any') # Null 값이 존재하는 행 제거
train_data.duplicated()
print(train_data.isnull().values.any()) # Null 값이 존재하는지 확인
train_data
```

```python
train_data = train_data.tolist()
```

#### 불용어 텍스트 파일 열기

```python
korean_stopwors_path = "korean_stopword.txt"

# 불용어 텍스트 파일 오픈
with open (korean_stopwors_path, encoding= 'utf8') as f : 
    stopwords = f.readlines()
stopwords = [x.strip() for x in stopwords]
```

#### Okt 명사 단위 분류

```python

# 형태소 분석기 OKT를 사용한 토큰화 작업 (다소 시간 소요)

# 명사 단위 분류 -- 3기 명사 추출함


okt = Okt()

tokenized_data = []
for sentence in (train_data):
    tokenized_sentence = okt.nouns(sentence) # 토큰화 # morphs 형태소 단위로 / .nouns()명사만 / phrases() 어절
    
    t_s = []
    for i in range(len(tokenized_sentence)):
        if len(tokenized_sentence[i]) > 1:
            t_s.append(tokenized_sentence[i])
        
    stopwords_removed_sentence = [word for word in t_s if not word in stopwords] # 불용어 제거
    tokenized_data.append(stopwords_removed_sentence)
```

```python
a =  sum(tokenized_data, [])
```

#### 추출된 키워드의 총 개수

```python
len(pd.DataFrame(a).drop_duplicates())
```

#### 빈도 수가 가장 높은 100개 키워드 개수 확인

```python
# # 키워드 갯수 확인
from collections import Counter

count = Counter(a)
tag_count = []
tags = []

for n, c in count.most_common(100):
    dics = {'tag': n, 'count': c}
    if len(dics['tag']) >= 2 and len(tags) <= 49:
        tag_count.append(dics)
        tags.append(dics['tag'])

for tag in tag_count:
    print(" {:<14}".format(tag['tag']), end='\t')
    print("{}".format(tag['count']))
    
```

#### 빈도수가 적은 단어 제거
- 빈도수가 5개 이하인 단어의 개수 확인

```python
count = Counter(a)

word_to_index = []
i = 0
for (word, frequency) in Counter(a).items():
    if frequency <= 5: # 빈도수가 적은 단어는 제외
        i += 1
        word_to_index.append(word)
# print(word_to_index)
len(word_to_index)
```

#### 빈도수가 5 이하인 단어 제거

```python
import tqdm
tokenized = tokenized_data.copy()
lesswords = word_to_index
lesswords = [x.strip() for x in lesswords]
tokenized_result = []
for i in tqdm.tqdm(tokenized):
    lesswords_removed_sentence = [word for word in i if not word in lesswords]
    tokenized_result.append(lesswords_removed_sentence)
```

```python
b =  sum(tokenized_result, [])
```

### Word2Vec

```python
from gensim.models import Word2Vec

model = Word2Vec(sentences = tokenized_result, vector_size = 200, window = 5, min_count = 5, sg = 0)
# sg - skip-gram : 1 / CBOW : 0
# workers = 1
```
