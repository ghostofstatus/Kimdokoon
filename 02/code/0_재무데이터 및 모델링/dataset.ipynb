### 1.3.1 2017 영업이익증가율, 순이익 증가율

```python
import pandas as pd
import numpy as np

df = pd.read_csv('./csv/ts2000/파생/총자본,영업이익,순이익 증가율.csv', encoding='cp949')

list = df['거래소코드'].unique()

for i in list:
    try:
            if df[df['거래소코드'] == i]['* (정상)영업손익(보고서기재)(IFRS연결)'].iloc[0] < 0 and df[df['거래소코드'] == i]['* (정상)영업손익(보고서기재)(IFRS연결)'].iloc[1] > 0 :  ## 적자에서 흑자로 넘어갈 경우
                if np.abs(df[df['거래소코드'] == i]['* (정상)영업손익(보고서기재)(IFRS연결)'].iloc[0]) <= np.abs(df[df['거래소코드'] == i]['* (정상)영업손익(보고서기재)(IFRS연결)'].iloc[1]) : ## 절대값 (당기)가 절대값 (전기)보다 클 경우
                    df.loc[df['거래소코드'] == i,'영업이익증가율']= df.groupby('거래소코드')['* (정상)영업손익(보고서기재)(IFRS연결)'].diff()/np.abs(df[df['거래소코드'] == i]['* (정상)영업손익(보고서기재)(IFRS연결)'].iloc[0])*100
                    ## 당기가 크면 전기로 나누어 줌
                else : 
                    df.loc[df['거래소코드'] == i,'영업이익증가율'] = df.groupby('거래소코드')['* (정상)영업손익(보고서기재)(IFRS연결)'].diff()/np.abs(df[df['거래소코드'] == i]['* (정상)영업손익(보고서기재)(IFRS연결)'].iloc[1])*100
                    ## 절대값의 크기가 전기가 더 크면 당기로 나누어줌
                    
            elif np.abs(df[df['거래소코드'] == i]['* (정상)영업손익(보고서기재)(IFRS연결)'].iloc[0]) == 0:
                df.loc[df['거래소코드'] == i,'영업이익증가율'] = 0
                
            else:
                df.loc[df['거래소코드'] == i,'영업이익증가율'] = df.groupby('거래소코드')['* (정상)영업손익(보고서기재)(IFRS연결)'].diff()/np.abs(df[df['거래소코드'] == i]['* (정상)영업손익(보고서기재)(IFRS연결)'].iloc[0])*100
                ## 전기와 당기의 부호가 같을 땐 원래의 식 (당기-전기)/전기
    except:
        pass

for i in list:
    try:
            if df[df['거래소코드'] == i]['당기순이익(손실)(IFRS연결)'].iloc[0] < 0 and df[df['거래소코드'] == i]['당기순이익(손실)(IFRS연결)'].iloc[1] > 0 :  ## 적자에서 흑자로 넘어갈 경우
                if np.abs(df[df['거래소코드'] == i]['당기순이익(손실)(IFRS연결)'].iloc[0]) <= np.abs(df[df['거래소코드'] == i]['당기순이익(손실)(IFRS연결)'].iloc[1]) : ## 절대값 (당기)가 절대값 (전기)보다 클 경우
                    df.loc[df['거래소코드'] == i,'순이익증가율']= df.groupby('거래소코드')['당기순이익(손실)(IFRS연결)'].diff()/np.abs(df[df['거래소코드'] == i]['당기순이익(손실)(IFRS연결)'].iloc[0])*100
                    ## 당기가 크면 전기로 나누어 줌
                else : 
                    df.loc[df['거래소코드'] == i,'순이익증가율'] = df.groupby('거래소코드')['당기순이익(손실)(IFRS연결)'].diff()/np.abs(df[df['거래소코드'] == i]['당기순이익(손실)(IFRS연결)'].iloc[1])*100
                    ## 절대값의 크기가 전기가 더 크면 당기로 나누어줌
            elif np.abs(df[df['거래소코드'] == i]['당기순이익(손실)(IFRS연결)'].iloc[0]) == 0:
                df.loc[df['거래소코드'] == i,'순이익증가율'] = 0
                
                
            else:
                df.loc[df['거래소코드'] == i,'순이익증가율'] = df.groupby('거래소코드')['당기순이익(손실)(IFRS연결)'].diff()/np.abs(df[df['거래소코드'] == i]['당기순이익(손실)(IFRS연결)'].iloc[0])*100
                ## 전기와 당기의 부호가 같을 땐 원래의 식 (당기-전기)/전기
    except:
        pass


df.to_csv('./csv/ts2000/파생/총자본,영업이익,순이익 증가율.csv', encoding='cp949')

```

```python
# 2017 연결 파생변수

import pandas as pd
import numpy as np


data2017_2_종합 = pd.read_csv('./csv/ts2000_2_2017_종합.csv', encoding= 'cp949')
data2016_2_종합 = pd.read_csv('./csv/ts2000_2_2017_종합전년도포함.csv', encoding= 'cp949')

data2017_2_종합.columns = ['회사명', '거래소코드', '회계년도', '매출액', '자산',
       '당기순이익', '현금및현금성자산', '이자비용',
       '부채', '자본', '영업손익',
       '판매비와관리비', '비유동자산', '이익잉여금']

data2016_2_종합.columns = ['회사명', '거래소코드', '회계년도', '자산', '부채', '자본', '현금및현금성자산']

data2016_2_종합_1 = data2016_2_종합.groupby('거래소코드').mean()
data2016_2_종합_1 = data2016_2_종합_1.reset_index()

data2017_2_종합_파생 = pd.DataFrame()
data2017_2_종합_파생['회사명'] = data2017_2_종합['회사명']
data2017_2_종합_파생['거래소코드'] = data2017_2_종합['거래소코드']
data2017_2_종합_파생['매출액'] = data2017_2_종합['매출액']
list = data2017_2_종합['거래소코드']

data2017_2_종합_파생['총자산영업이익율']= ''
for i in list : 
       data2017_2_종합_파생.loc[data2017_2_종합_파생['거래소코드']==i,'총자산영업이익율'] = data2017_2_종합.loc[data2017_2_종합['거래소코드']==i,'영업손익'].values/data2016_2_종합_1.loc[data2016_2_종합_1['거래소코드']==i,'자산'].values # 수정 필요 (시점이 다른 것들이 존재함)

data2017_2_종합_파생['총자산부채비율'] = data2017_2_종합['부채']/data2017_2_종합['자산']
data2017_2_종합_파생['log총자산'] = np.log10(data2017_2_종합['자산'])

data2017_2_종합_파생['순이익률']= ''
for i in list : 
       data2017_2_종합_파생.loc[data2017_2_종합_파생['거래소코드']==i,'순이익률'] = data2017_2_종합.loc[data2017_2_종합['거래소코드'] == i,'당기순이익'].values/data2016_2_종합_1.loc[data2016_2_종합_1['거래소코드']==i,'자산'].values # 수정 필요

data2017_2_종합_파생['현금비율'] = data2017_2_종합['현금및현금성자산']/data2017_2_종합['자산']

data2017_2_종합_파생['이자비용/부채']= ''
for i in list : 
       data2017_2_종합_파생.loc[data2017_2_종합_파생['거래소코드'] == i,'이자비용/부채'] = data2017_2_종합.loc[data2017_2_종합['거래소코드'] == i,'이자비용'].values/data2016_2_종합_1.loc[data2016_2_종합_1['거래소코드'] == i,'부채'].values # 수정 필요

data2017_2_종합_파생['자기자본비율'] = data2017_2_종합['자본']/data2017_2_종합['자산']
data2017_2_종합_파생['고정비율'] = data2017_2_종합['비유동자산']/data2017_2_종합['자산'] 
data2017_2_종합_파생['매출액대판매관리비'] = data2017_2_종합['판매비와관리비']/data2017_2_종합['매출액'] 
data2017_2_종합_파생['유보이익대총자산비율'] = data2017_2_종합['이익잉여금']/data2017_2_종합['자산'] 
data2017_2_종합_파생
```

```python
# 데이터 종합하기
data2017_종합 = pd.read_csv('./csv/ts2000_2017_종합.csv', encoding= 'cp949')
data2019_종합 = pd.read_csv('./csv/ts2000_2019_종합.csv', encoding= 'cp949')

data2017_업종 = pd.read_csv('./csv/산업분류.csv' , encoding= 'cp949')
data2019_업종 = pd.read_csv('./csv/산업분류2019.csv' , encoding= 'cp949')

data2017_업종.drop('회계년도', axis=1,inplace=True)
data2019_업종.drop('회계년도', axis=1,inplace=True)

data2017_종합.drop(['정상영업이익증가율(IFRS연결)','순이익증가율(IFRS연결)'], axis= 1, inplace=True)
data2019_종합.drop(['정상영업이익증가율(IFRS연결)','순이익증가율(IFRS연결)'], axis= 1, inplace=True)

df = pd.read_csv('./csv/ts2000/파생/총자본,영업이익,순이익 증가율.csv', encoding='cp949')
df2 = pd.read_csv('./csv/ts2000/파생/총자본,영업이익,순이익 증가율2019.csv', encoding='cp949')

df = df[(df['회계년도'] == '17-Dec') | (df['회계년도'] == '17-Jun') | (df['회계년도'] == '17-Mar')]
df2 = df2[(df2['회계년도'] == '19-Dec') | (df2['회계년도'] == '19-Jun') | (df2['회계년도'] == '19-Mar')]

df = df[['회사명', '거래소코드', '영업이익증가율', '순이익증가율']]
df2 = df2[['회사명', '거래소코드', '영업이익증가율', '순이익증가율']]

data2017_total = pd.merge(data2017_종합, df, on = ['회사명','거래소코드'], how = 'left')
data2019_total = pd.merge(data2019_종합, df2, on = ['회사명','거래소코드'], how = 'left')

data2017_total = pd.merge(data2017_total, data2017_2_종합_파생, on = ['회사명','거래소코드'], how = 'left')
data2019_total = pd.merge(data2019_total, data2019_2_종합_파생, on = ['회사명','거래소코드'], how = 'left')

data2017_total = pd.merge(data2017_total, data2017_업종, on = ['회사명','거래소코드'], how = 'left')
data2019_total = pd.merge(data2019_total, data2019_업종, on = ['회사명','거래소코드'], how = 'left')
```

```python
import pandas as pd

list2017 = pd.read_csv('./csv/제외기업2017.csv', encoding='cp949')
list2019 = pd.read_csv('./csv/제외기업2019.csv', encoding='cp949')
list = pd.read_excel('./csv/관리종목.xlsx')

list2017 = list2017['거래소코드']
list2019 = list2019['거래소코드']
list = list['종목코드']

for i in list2017 :
    index = data2017_total[data2017_total['거래소코드'] == i].index
    data2017_total.drop(index, axis=0, inplace=True)
    
for i in list2019 :
    index = data2019_total[data2019_total['거래소코드'] == i].index
    data2019_total.drop(index, axis=0, inplace=True)
    
for i in list:
    index = data2017_total[data2017_total['거래소코드'] == i].index
    index1 = data2019_total[data2019_total['거래소코드'] == i].index
    data2017_total.drop(index, axis=0, inplace=True)
    data2019_total.drop(index1, axis=0, inplace=True)

```

```python
# 업종 인코딩
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
encoder.fit(data2017_total['통계청 한국표준산업분류 10차(대분류)'])
```

```python
data2017_total['업종']=encoder.transform(data2017_total['통계청 한국표준산업분류 10차(대분류)'])
data2019_total['업종']=encoder.transform(data2019_total['통계청 한국표준산업분류 10차(대분류)'])
```

```python
## 기업 규모로 나누기

data2017_total['규모'] = ''
data2019_total['규모'] = ''

for i in range(0, len(data2017_total) , 1) :
    if data2017_total.loc[i,'매출액'] > 100000000 :
        data2017_total.loc[i,'규모'] = 1
    else :
        data2017_total.loc[i,'규모'] = 0
for i in range(0, len(data2019_total) , 1) :
    if data2019_total.loc[i,'매출액'] > 100000000 :
        data2019_total.loc[i,'규모'] = 1
    else : 
        data2019_total.loc[i,'규모'] = 0
```

```python
data2017_total['층화'] = ''
data2019_total['층화'] = ''
for i in range (0, 2, 1):
    for j in range(0, 3, 1):
        if i == 0 :
            data2017_total.loc[(data2017_total['업종'] == j) & (data2017_total['규모'] == i),'층화'] = j
            data2019_total.loc[(data2019_total['업종'] == j) & (data2019_total['규모'] == i),'층화'] = j
        else:
            data2017_total.loc[(data2017_total['업종'] == j) & (data2017_total['규모'] == i),'층화'] = j+3
            data2019_total.loc[(data2019_total['업종'] == j) & (data2019_total['규모'] == i),'층화'] = j+3
```

```python
data2017_total.drop('매출액',axis=1, inplace=True)
data2019_total.drop('매출액',axis=1, inplace=True)

data2017_total.drop('자기자본비율',axis=1, inplace=True)
data2019_total.drop('자기자본비율',axis=1, inplace=True)
```
