### 관리종목지정 기업 Train dataset

```python
train_total = pd.read_csv('ts2000_부실_total_trainset.csv', encoding='cp949')
# train_total.info()
```

##### 연결재무제표 파생변수

```python
train_total['총자산부채비율'] = train_total['부채(*)(IFRS연결)'] / train_total['자산(*)(IFRS연결)']
train_total['순이익률'] = train_total['당기순이익(손실)(IFRS연결)'] / train_total['자산(*)(IFRS연결)']
train_total['현금비율'] = train_total['현금및현금성자산(*)(IFRS연결)'] / train_total['자산(*)(IFRS연결)']
train_total['금융비용/부채 비율'] = train_total['이자비용(IFRS연결)'] / train_total['부채(*)(IFRS연결)']
train_total['유보이익대총자산비율'] = train_total['이익잉여금(결손금)(*)(IFRS연결)'] / train_total['자산(*)(IFRS연결)']
train_total['자기자본비율'] = train_total['자본(*)(IFRS연결)'] / train_total['자산(*)(IFRS연결)']
train_total['판매관리비대매출액'] = train_total['판매비와 관리비(물류원가 등 포함)(*)(IFRS연결)'] / train_total['매출액(수익)(*)(IFRS연결)']
train_total['고정비율'] = train_total['비유동자산(*)(IFRS연결)'] / train_total['자산(*)(IFRS연결)']
train_total['총자산영업이익율'] = train_total['* (정상)영업손익(보고서기재)(IFRS연결)'] / train_total['자산(*)(IFRS연결)']
train_total['log총자산'] = np.log10(train_total['자산(*)(IFRS연결)'])
```

##### 개별재무제표 파생변수

```python
train_total['총자산부채비율2'] = train_total['부채(*)(IFRS)'] / train_total['자산(*)(IFRS)']
train_total['순이익률2'] = train_total['당기순이익(손실)(IFRS)'] / train_total['자산(*)(IFRS)']
train_total['현금비율2'] = train_total['현금및현금성자산(*)(IFRS)'] / train_total['자산(*)(IFRS)']
train_total['금융비용/부채 비율2'] = train_total['이자비용(IFRS)'] / train_total['부채(*)(IFRS)']
train_total['유보이익대총자산비율2'] = train_total['이익잉여금(결손금)(*)(IFRS)'] / train_total['자산(*)(IFRS)']
train_total['자기자본비율2'] = train_total['자본(*)(IFRS)'] / train_total['자산(*)(IFRS)']
train_total['판매관리비대매출액2'] = train_total['판매비와 관리비(물류원가 등 포함)(*)(IFRS)'] / train_total['매출액(수익)(*)(IFRS)']
train_total['고정비율2'] = train_total['비유동자산(*)(IFRS)'] / train_total['자산(*)(IFRS)']
train_total['총자산영업이익율2'] = train_total['* (정상)영업손익(보고서기재)(IFRS)'] / train_total['자산(*)(IFRS)']
train_total['log총자산2'] = np.log10(train_total['자산(*)(IFRS)'])
```

##### 연결재무제표 2차 파생변수

```python
train_total['총자본증가율'] = train_total.groupby('회사명')['자본(*)(IFRS연결)'].pct_change() * 100
train_total['유동자산증가율'] = train_total.groupby('회사명')['유동자산(*)(IFRS연결)'].pct_change() * 100
train_total['매출액증가율'] = train_total.groupby('회사명')['매출액(수익)(*)(IFRS연결)'].pct_change() * 100
train_total['차입금의존도'] = (train_total['장기차입금(*)(IFRS연결)'] + train_total['단기차입금(*)(IFRS연결)'] + train_total['유동성장기부채(*)(IFRS연결)']) / train_total['자산(*)(IFRS연결)'] 
train_total['매출총이익률'] = train_total['매출총이익(손실)(IFRS연결)'] / train_total['매출액(수익)(*)(IFRS연결)'] * 100
train_total['재고자산회전율'] = train_total['매출액(수익)(*)(IFRS연결)'] / train_total['재고자산(*)(IFRS연결)']
train_total['매출액대매출원가'] = train_total['매출원가(*)(IFRS연결)'] / train_total['매출액(수익)(*)(IFRS연결)'] * 100
```
