```python
import pandas as pd

df = pd.read_excel('./data/관리종목.xlsx')
df['시간'] = pd.to_datetime(df['시간'])
df = df.sort_values('시간')
df = df.set_index(keys='시간')
df = df.reset_index()
df.duplicated('회사명').sum() # 관리종목으로 중복 지정된 회사 29개
#df[df['회사명'] == "UCI"]
#df = df.drop_duplicates(['회사명'], keep = 'first') #가장 먼저 지정된 시점만 남기고 중복 데이터 삭제
#df[df['회사명'] == "UCI"]
df['시간'] = df['시간'].dt.strftime("%Y.%m.%d")
df['시간'] = pd.to_datetime(df['시간'])

df = df[['시간', '회사명', '종목코드', '공시제목', '제출인']]
# df.to_csv('./data/관리종목.csv')
df
```

```python
ove = df[df.duplicated('회사명')]
ove = ove.sort_values('회사명')
ove['회사명'].unique()
```

```python
from dateutil.relativedelta import relativedelta
# 6개월 전 시간 추출
new = []
for i in range(0, 244):
    n = df.loc[i]['시간'] - relativedelta(months=6)
    new.append(n)
```

```python
df['기간'] = pd.DataFrame(new)
```

```python
df = df[['시간', '회사명', '종목코드', '공시제목', '제출인', '기간']]
```
